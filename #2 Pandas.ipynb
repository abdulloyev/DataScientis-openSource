## *DataFrameda qayta indekslash uchun .index funksiyasidan foydalaniladi.*

```python
data = {
    'Model': ['Mercedes', 'BMW', 'Tesla', 'Toyota', 'Volkswagen', 'Volvo'],
    'Year': [2010, 2012, 2014, 2016, 2018, 2019],
    'Temp': [1.34, 1.89, 2.11, 2.21, 2.54, 2.98],
    'Price': [4.75, 6.54, 8.74, 6.32, 5.21, 3.98]
}

df = pd.DataFrame(data)
df
```

## *df nomli DataFrameni indekslarini oʻzgartiramiz*

```python
df.index=[34, 45, 123, 56, 67, 3]
df
```

## *.reindex funksiyasi yordamida indekslar oʻrnini almashtirish mumkin.*

```python
df.reindex([56, 3, 67, 34])
```

## *reindex() metodi yangi indekslarni qo‘shadi, agar yangi indekslarda mos keluvchi qator bo‘lmasa, ffill metodi orqali bo‘sh joylar oldingi qiymatlar bilan to‘ldiriladi.*

```python
df = df.reindex(index=range(71, 79), method='ffill')
```

## *Ustunlarni ham qayta tartiblash mumkin:*

```python
df = df.reindex(columns=['Price', 'Year', 'Wrong'])
df
```

## *DataFramedan elementlarni oʻchirish. .drop()DataFramedan biror qator yoki ustunni oʻchirish uchun qator (ustun) indeksini bilish kifoya:*

```python
df2 = df.drop(3)
df2
```

## *axis=1 – ustun bo‘ylab o‘chirishni bildiradi (satirlar uchun axis=0 ishlatiladi).*

```python
df2 = df.drop('Year', axis=1)
df2
```

## *.drop va .reindex kabi funksiyalarni qoʻllaganimizda asl DataFrame oʻzgarmaydi, lekin buni maxsus “inplace” parametri yordamida oʻzgarishlarni saqlashimiz mumkin. Bu parametr standart False qiymatga ega boʻladi.*

```python
df.drop('Year', axis=1, inplace=False)
```

## Pandas kutubxonasida arifmetik amallar

## *amallar bajarish uchun 2 ta “s1” va “s2” Series obyektlarini yaratib olamiz.*

```python
s1=pd.Series([12,15,6,9.12,11],index=['a','b','c','d','e'])
s2=pd.Series([8,6.5,4,54,3],index=['a','b','h','t','d'])
```

## *s1+s2 amalini natijasini koʻrishimiz mumkin.*

```python
s1+s2
```

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame(
    np.arange(9).reshape(3,3),
    columns=list('bcd'),
    index=['Apelsin', 'Banan', 'Gilos']
)

df2 = pd.DataFrame(
    np.arange(12).reshape(4,3),
    columns=list('abc'),
    index=['Apelsin', 'Banan', 'Olcha', 'Anjir']
)
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

## *Yuqoridagi qoʻshish amalini .add() metodi yordamida ham bajarish mumkin.*

```python
df1.add(df2)
```

## *Bunda maxsus fill_value parametri yordamida mos tushuvchi indekslardagi NaN qiymatlarni boshqa qiymat bilan almashtirish mumkin. Bu usulda yakuniy natijada ikkala df da ham mavjud boʻlmagan indekslarda NaN qiymati qoladi.*

```python
df_result = df2.add(df1, fill_value=0)
df_result
```
