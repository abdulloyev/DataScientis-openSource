# Kalssik dastur

  **KLASSIK DASTUR**: Odatda, biz dasturchi sifatida biror vazifani bajaruvchi
dasturlar yozamiz. Bu dasturlarni yozishdan avval qanday qilib shu vazifani bajarish

#### **`make_regression() `**: sintetik (sun'iy) regressiya ma'lumotlarini yaratish uchun ishlatiladi

* **`n_samples=40 `** 40 ta namuna (data point) yaratishni bildiradi. Ya'ni, ma'lumotlar to'plamida 40 ta nuqta bo'ladi.

* **`n_features=1 `** Har bir namunada 1 ta xususiyat (feature) bo'ladi.

* **`noise=10`** Ma'lumotlarga qo'shiladigan shovqin (noise) miqdori. Bu qiymat qanchalik katta bo'lsa, ma'lumotlar shunchalik tarqoq bo'ladi.

* **`random_state=42`** Random sonlarni generatsiya qilish uchun boshlang'ich qiymat (seed).



```python
# DATA
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
```

```python
# data[0] - bu x_data (mustaqil o'zgaruvchi), data[1] - bu y_data (bog'liq o'zgaruvchi).
x_data, y_data = data[0], data[1]
```

```python
# yuzoridagi natigani chizamiz
plt.scatter(x_data, y_data)
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('Regression Data')
plt.show()
```

# Regressiya tushunchasini bashorat chizig'i

#### Bu kodning asosiy maqsadi ***regressiya tushunchasini*** tushuntirish va oddiy ***model yordamida bashorat qilishni*** ko'rsatishdir.

```python
def classic_program(x):
  return x * 70

y_pred = [classic_program(x) for x in x_data]

# Plotting

plt.scatter(x_data, y_data, label="Haqiqiy data sets")
plt.plot(x_data, y_pred, 'red', label="Dastur")
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend() # grafikdagi chixmalar nomi legenda ko'rsatadi
plt.title('Regression Data (x * 70)')
plt.show()
```

## **`Vazn qo'shilgan regressiya modeli`**

#### Bu kodning asosiy maqsadi vazn ***(weight) qo'shilgan regressiya modeli*** yordamida ma'lumotlarni tahlil qilish va bashorat qilishni ko'rsatishdir

```python
# DATA

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

def weight_assigend_program(x, w):
  return x * w * 70
w = 1.3
y_pred = [weight_assigend_program(x, w) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label="Haqiqiy data sets")
plt.plot(x_data, y_pred, color="red", label="Vazinli dastur")

plt.xlabel("X")
plt.ylabel("Y")
plt.title(f'{w} * x * 70')
plt.legend()
plt.show()
```

Dastur endi Model deb yuritilinmoqda. Sababi model ham o’zi dastur hisoblanib,
dasturdan farqli o’laroq endi parameterlarga ham ega bo’ladi.

# MACHINE LEARNING MODELI

* **Kirish ma'lumotlari:** Bu qismda modelga kirish ma'lumotlari (input data) beriladi. Masalan, `x_data` va `y_data` kabi ma'lumotlar.

* **Model:** Bu qismda Machine Learning modeli yaratiladi. Model ma'lumotlarni qabul qilib, ularni tahlil qiladi va bashorat qiladi.

* **Natija:** Modelning ishlashi natijasida hosil bo'lgan bashoratlar (predictions) ko'rsatiladi. Masalan, `y_pred` qiymatlari.

* **Ko'rsatgich:** Modelning ishlashini baholash uchun ko'rsatgichlar (metrics) ishlatiladi. Masalan, aniqlik (`accuracy`), xato (`error`) yoki boshqa metrikalar.

* **Vazn:** Modelning parametrlari (`weights`) ko'rsatiladi. Bu parametrlar modelni o'qitish jarayonida yangilanadi.

* **Yangilash:** Modelni o'qitish jarayonida vaznlar (weights) yangilanadi. Bu jarayon ma'lumotlar asosida modelni yaxshilash uchun amalga oshiriladi.

image.png

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

# DATA

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# regressiya
regression = LinearRegression().fit(x_data, y_data)
y_pred = regression.predict(x_data)

# ploting
plt.scatter(x_data, y_data, label="Haqiqiy data sets")
plt.plot(x_data, y_pred, "red", label="ML datur")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
equation = f'{regression.coef_[0]:.3f}x+{regression.intercept_:.3f}'
print(equation)
plt.show()
```
