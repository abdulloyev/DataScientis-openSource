## Kerakli kutubxonalarni va DataFrameni yaratib olamiz!

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

```python
# X va Y
X_col = [10, 20, 30, 40, 50]
Y_col = [6000, 7000, 8000, 9000, 10000]
```

```python
# DataFrame yaratamiz
df = pd.DataFrame({
    "X": X_col,
    "Y": Y_col
})

df
```

## Modelni yaratib olamiz

```python
model = LinearRegression()

# Chiziqli regressiya modelini qurish
X = df[["X"]] #Seris
y = df["Y"]

model.fit(X, y)
```

## Regressiya tenglamasi

```python
# model parametrlari
a = model.intercept_
b = model.coef_[0]

T = f"y = {a:.0f} + {b:.0f}x"
T
```

## Basharat qilamiz

```python
df['predict'] = model.predict(X)
df
```

## Modelni baholaymiz

```python
r2 = r2_score(y, df['predict'])
print(f"R2 = {r2:.2f}")
```

## Grafik chizish

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['X'], df['Y'], label='Haqiqiy qiymatlar')
plt.plot(df['X'], df['predict'], color='red', label='Regressiya chizig\'i')

plt.title('Harorat va Energiya Iste\'moli')
plt.xlabel('Harorat')
plt.ylabel('Energiya iste\'moli')
plt.show()
```

## Yangi qiymatlar uchun bashorat

```python
new_df = pd.DataFrame({'X': [15, 25, 35, 45]})
new_df['predict'] = model.predict(new_df[['X']])

new_df
```
